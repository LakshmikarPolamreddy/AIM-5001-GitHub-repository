### Web scraping of fines data from html page for each violation code

```python
# pd.read_html will extract the list of all the tables in the webpage

tables = pd.read_html("https://www.nyc.gov/site/finance/vehicles/services-violation-codes.page", header = 0)
tables[0] # printing the first table
```

```python
len(tables) # Printing the number of tables available on the web page
```

```python
# Merging all the tables vertically to generate a data frame of violation codes

violation_code = pd.concat([tables[0], tables[1], tables[2], tables[3], tables[4]], ignore_index=True)
violation_code
```

```python
# For our analysys, we assume fines are same throughout NYC and so ignoring column 3 which is specific to Manhattan 96thst and below
violation_code.drop(['Manhattan96th St. & below'], axis=1, inplace=True)
```

```python
# Renaming the last column to Fines in USD

violation_code.rename(columns = {'All Other Areas' : 'Fines in USD'}, inplace = True)
violation_code.head()
```

```python
#violation_code['Fines in USD'] = violation_code['Fines in USD'].replace('$','')

violation_code['Fines in USD'] = violation_code['Fines in USD'].str.replace("$",'')
violation_code
```

```python
# Replacing the text and special characters with values in USD
# Replacing blank values with NaN

violation_code['Fines in USD'] = violation_code['Fines in USD'].replace("1st Offense - 502nd Offense - 1003rd Offense - 1504th Offense - 200Each subsequent offense - 250",'50')
violation_code['Fines in USD'] = violation_code['Fines in USD'].replace("1st Offense - 2502nd Offense - 500",'50')
violation_code['Fines in USD'] = violation_code['Fines in USD'].replace("(Regular Tow, plus violation fine) $200 (Heavy Tow, plus violation fine)",'50')
violation_code['Fines in USD'] = violation_code['Fines in USD'].replace("--",'NaN')
violation_code['Fines in USD'] = violation_code['Fines in USD'].replace("65**",65)
violation_code['Fines in USD'] = violation_code['Fines in USD'].replace("(Regular Tow, plus violation fine) 200 (Heavy Tow, plus violation fine)",200)
violation_code['Fines in USD'] = violation_code['Fines in USD'].replace("vary",'NaN')
violation_code['Fines in USD'] = violation_code['Fines in USD'].astype(float)
violation_code
```

```python
violation_code = violation_code
df1 = violation_code.copy()
df1 = df1.drop((df1.loc[df1["CODE"].isnull()]).index).reset_index().drop("index",axis=1)

for i in range(0,len(df1)):
    df1["CODE"][i] = str(df1["CODE"][i]).split("-")[0]
    df1["CODE"][i] = str(df1["CODE"][i]).split(".")[0]
    
df1
```

- Suburban vehicles and 4 door sedan vehicles are involved in 70% of the parking violations and vehicle brands such as Honda, Toyota, Ford and Nissan accounted for 41.3% of the parking violations

- The highest number of parking violations accounting for 23.9% occurred in Manhattan borough

- 75% of the vehicles involved in parking violations belong to NY state indicating that most of the violators are from NY state.

- Summer months have witnessed more number of parking violations when compared to other months

- The share of unregistered vehicles involved in illegal parking is just 2%

- Driver distraction is the major reason for accidents in NYC

- The highest number of accidents accounting for 31.6% occurred in Brooklyn borough, followed by Queens

- The highest number of collisions occurred on Broadway steet

- October month accounts for the highest number of accidents while the least number of accidents happened in February

The major finding of our analysis is : though the highest number of parking violations happened in Manhattan borough, the highest number of accidents did not occur there. Instead, Brooklyn borough witnessed the largest share of accidents followed by Queens. We could not establish here that the parking violations are responsible for the collisions happening in NYC.
